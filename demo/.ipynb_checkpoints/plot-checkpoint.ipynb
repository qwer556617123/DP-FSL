```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
```

```python
name = ['Fedavg','Fedadam']
```

```python
train_1_loss = {}
train_1_sec_acc = {}
train_1_win_acc = {}
train_1_f1 = {}
train_1_recall = {}
train_1_precision = {}

hyper_param = '_132'

for i in name:
    string = '../data/result/' + i + hyper_param +  '_1' + '_train_loss.csv'
    df = pd.read_csv(string, header=None)
    df = np.mean(np.array(df), axis = 0)
    train_1_loss[i] = df.reshape(-1)
    del df
    
for i in name:
    string = '../data/result/' + i + hyper_param +  '_1' + '_train_sec_acc.csv'
    df = pd.read_csv(string, header=None)
    df = np.mean(np.array(df), axis = 0)
    train_1_sec_acc[i] = df.reshape(-1)
    del df
    
for i in name:
    string = '../data/result/' + i + hyper_param +  '_1' + '_train_win_acc.csv'
    df = pd.read_csv(string, header=None)
    df = np.mean(np.array(df), axis = 0)
    train_1_win_acc[i] = df.reshape(-1)
    del df
    
for i in name:
    string = '../data/result/' + i + hyper_param +  '_1' + '_train_f1.csv'
    df = pd.read_csv(string, header=None)
    df = np.mean(np.array(df), axis = 0)
    train_1_f1[i] = df.reshape(-1)
    del df
    
for i in name:
    string = '../data/result/' + i + hyper_param +  '_1' + '_train_recall.csv'
    df = pd.read_csv(string, header=None)
    df = np.mean(np.array(df), axis = 0)
    train_1_recall[i] = df.reshape(-1)
    del df
    
for i in name:
    string = '../data/result/' + i + hyper_param +  '_1' + '_train_precision.csv'
    df = pd.read_csv(string, header=None)
    df = np.mean(np.array(df), axis = 0)
    train_1_precision[i] = df.reshape(-1)
    del df
```

```python
show_label_start = 0
show_label_end = 500
```

```python
def lbound(x):
    a = []
    for i in x:
        a.append(max(0.0,i))
    return a
def hbound(x):
    a = []
    for i in x:
        a.append(min(1.0,i))
    return a
```

# Train

```python
for i in name:
    x = np.arange(show_label_start,show_label_end)
    y = train_1_loss[i][show_label_start+1:show_label_end+1]
    y2 = train_10_loss[i][show_label_start+1:show_label_end+1]
    
    print(i)
    print('e = 1 ', y[-1])
    print('e = 10 ', y2[-1])
    
    c = interp1d(x,y, kind = "cubic")
    c2 = interp1d(x,y2, kind = "cubic")
    x = np.linspace(x.min(), x.max(), int((show_label_end-show_label_start)/5))
    y = c(x)
    y2 = c2(x)
    y = lbound(y)
    y2 = lbound(y2)
    plt.plot(x,y, label = i + ' e = 1')
    plt.plot(x,y2, label = i + ' e = 10')
    
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training Loss')
plt.legend()
#plt.legend(bbox_to_anchor=(1.05, 0), loc=3)
# plt.savefig('../data/original_train_loss.svg', bbox_inches='tight', format="svg")
plt.show()
```
